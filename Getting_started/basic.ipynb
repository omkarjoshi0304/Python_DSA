```python
n = int (input("Enter a number: "))
sum = 0
count = 1
while count<=n:
    sum = sum + count
    count += 1
print (f"Sum of {n} natural number is : ",sum)
```

```python
n = int (input ("Enter a number : "))
sum = 0

for i in range ( n+1 ):
    sum = sum+i
print (f"Sum of {n} natural number is : ",sum)
```

```python
## exmaple : prime number between 1 and 100

for i in range (1,101):
    if i>1:
        for j in range (2,i):
            if i%j==0:
                break
        else:
            print(i)
```
