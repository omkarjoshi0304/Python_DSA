### Create a TODO list

```python
to_do_list = ["Buy Groceries","Clean the house", "Pay bills"]

# adding things in the todo list

to_do_list.append("office meetings")
to_do_list.append("Go for a run")

# removing things from the todo list

to_do_list.remove("Buy Groceries")

# checking if a task is in the list

if "Pay bills" in to_do_list:
    print("DO not forget to pay your bills")

print ("To do list reminder")

for task in to_do_list:
    print(f"{task}")

```

### Organizing the Student Grades

```python
# Organizing the students grades

grades = [85, 92, 78 ,88, 96]

# adding the grades 

grades.append(87)

# calculating the avg

average_grade = sum(grades) / len(grades)
print(f"Average grades: {average_grade}")

# Highest and the lowest grades

highest_grade = max(grades)
print(f"The higest grade you have scored is: {highest_grade}")

lowest_grade = min(grades)
print(f"The lowest grade you have scored is: {lowest_grade}")
```

### Manging the inventory

```python
inventory = ["apple","grapes","bananas","oranges"]

# adding an item 

inventory.append("mango")

# removing the item from the inventory

inventory.remove("apple")

# checking if the item is present or not

item = "mango"

if item in inventory:
    print(f"{item} is present in the stock")
else:
    print(f"{item} is out of stock")
```
