```python
# create a set

my_set = {1,2,3,5,6}
print(my_set)
```

```python
# type casting of a set

set1 = set([1,2,3,4,5,6,7])
print(set1)
```

```python
#only unique elements 
set2 = {1,2,3,4,4,4}
print (set2)
```

```python
set2.add(5)
print (set2)
```

```python
set2.remove(2)
print(set2)
```

```python
## set memebership test
set3 = {1,2,3,4,5,5,6,7,8}
print(2 in set3)
print(9 in set3)
```

```python
## mathematical operations

set4 = {1,2,3,4,5,6}
set5 = {4,5,6,7,8,9}

## union (combine)

union_set= set4.union(set5)
print(union_set)

## Intersection (commmon)

intersection_set = set4.intersection(set5)
print (intersection_set)

```

```python
## set methods

set4 = {1,2,3,4,5,6}
set5 = {4,5,6,7,8,9}

## is subset checkin if  any comman element is there or not

print(set4.issubset(set5))
```
